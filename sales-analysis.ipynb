# Merge Datasets:


```python
# Merge train_data with stores_data on 'Store'
merged_data = pd.merge(train_data, stores_data, on='Store', how='left')

# Merge merged_data with features_data on 'Store' and 'Date'
merged_data = pd.merge(merged_data, features_data, on=['Store', 'Date'], how='left')

# Display the first few rows of the merged dataset
print("Merged Data:")
merged_data.head()
```

```python

# Assuming the merged data is stored in a DataFrame called 'merged_data'

# Select numerical variables
numerical_vars = ['Weekly_Sales', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

# Compute descriptive statistics
descriptive_stats = merged_data[numerical_vars].describe()

# Print the descriptive statistics
print("Descriptive Statistics for Numerical Variables:")
descriptive_stats
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
correlation_matrix = merged_data[numerical_vars].corr()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

stores analysis

```python
import seaborn as sns

# Group data by store type and calculate summary statistics for weekly sales
sales_by_type = merged_data.groupby(['Type'])['Weekly_Sales'].describe()

# Display summary statistics
print("Summary Statistics for Weekly Sales by Store Type:")
print(sales_by_type)

# Create a bar plot to visualize sales performance across different store types
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_type.index, y=sales_by_type['mean'], palette='viridis')
plt.title('Average Weekly Sales by Store Type')
plt.xlabel('Store Type')
plt.ylabel('Average Weekly Sales')
plt.show()
```

Store size analysis

```python
# Group data by store size and calculate summary statistics for weekly sales
sales_by_size = merged_data.groupby('Size')['Weekly_Sales'].describe()


# Create a scatter plot to visualize the relationship between store size and weekly sales
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Size'], merged_data['Weekly_Sales'], alpha=0.5)
plt.title('Weekly Sales vs. Store Size')
plt.xlabel('Store Size')
plt.ylabel('Weekly Sales')
plt.show()
```

Seasonal Analysis

```python
# Extract month and year from the 'Date' column
merged_data['Year'] = pd.to_datetime(merged_data['Date']).dt.year
merged_data['Month'] = pd.to_datetime(merged_data['Date']).dt.month

# Group data by year and month, and calculate summary statistics for weekly sales
sales_by_month = merged_data.groupby(['Year', 'Month'])['Weekly_Sales'].sum()

# Reset index to make the grouped data accessible for visualization
sales_by_month = sales_by_month.reset_index()

# Visualize monthly sales trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_by_month, x='Month', y='Weekly_Sales', hue='Year', marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Weekly Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

```python
# Encode categorical variable 'IsHoliday' using numerical format (0 for False, 1 for True)
train_data['IsHoliday'] = train_data['IsHoliday'].astype(int)

# Display the first few rows of the preprocessed training data
print("Preprocessed Training Data:")
print(train_data.head())
```

```python
# Define features (X) and target variable (y)
x = train_data.drop(['Weekly_Sales', 'Date'], axis=1)
y = train_data['Weekly_Sales']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(x_train, y_train)


# Save the trained model for future use
from joblib import dump

# Save the trained model to a file
dump(model, 'linear_regression_model.joblib')

print("Linear Regression model training completed and saved successfully!")
```

```python
from sklearn.metrics import r2_score

# Make predictions on the training data
y_pred_train = model.predict(x_test)

# Calculate Mean Squared Error (MSE) on the training data
mse_train = r2_score(y_test, y_pred_train)
print("Root Mean Squared Error (RMSE) on training data:", mse_train)
```
